# Linear Regression with Scikit-Learn

```python
from sklearn.linear_model import LinearRegression
```

```python
# we see high correlation between RM (no of rooms) and 
# MEDV (Median value by which owner-occupied home)
# so RM is predictor and MEDV is response

X=df['RM'].values
```

```python
X
```

```python
X=df['RM'].values.reshape(-1,1)
X
```

```python
Y=df['MEDV'].values
Y
```

```python
model=LinearRegression()
```

```python
model.fit(X,Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(12,7))
sns.regplot(X,Y)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.regplot(X,Y)
plt.xlabel('average number of rooms per dwelling')
plt.ylabel("Median value of owner-occupied homes in $1000's")
plt.show();
```
